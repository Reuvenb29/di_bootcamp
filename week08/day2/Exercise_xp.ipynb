```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/di_bootcamp_resources/week_8/titanic dataset/train.csv')
display(df.head())
```

Exercise 2

```python
missing_values = df.isnull().sum()
print("Columns with missing values:")
display(missing_values[missing_values > 0])
```

```python
# Impute missing 'Age' values with the mean
df['Age'] = df['Age'].fillna(df['Age'].mean())
display(df.head())
```

```python
# Impute missing 'Embarked' values with the mode
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
display(df.head())
```

```python
# Calculate IQR for 'Fare'
Q1_fare = df['Fare'].quantile(0.25)
Q3_fare = df['Fare'].quantile(0.75)
IQR_fare = Q3_fare - Q1_fare

# Define bounds for 'Fare' outliers
lower_bound_fare = Q1_fare - 1.5 * IQR_fare
upper_bound_fare = Q3_fare + 1.5 * IQR_fare

# Identify outliers in 'Fare'
outliers_fare = df[(df['Fare'] < lower_bound_fare) | (df['Fare'] > upper_bound_fare)]
print(f"Number of outliers in 'Fare' using IQR: {len(outliers_fare)}")
display(outliers_fare.head())

# Calculate IQR for 'Age'
Q1_age = df['Age'].quantile(0.25)
Q3_age = df['Age'].quantile(0.75)
IQR_age = Q3_age - Q1_age

# Define bounds for 'Age' outliers
lower_bound_age = Q1_age - 1.5 * IQR_age
upper_bound_age = Q3_age + 1.5 * IQR_age

# Identify outliers in 'Age'
outliers_age = df[(df['Age'] < lower_bound_age) | (df['Age'] > upper_bound_age)]
print(f"\nNumber of outliers in 'Age' using IQR: {len(outliers_age)}")
display(outliers_age.head())
```
