```python
# prompt: i wan to imprt a data set, heres the file path /content/drive/MyDrive/Colab Notebooks/Di Bootcamp files/datascience_salaries.csv

import pandas as pd

file_path = '/content/drive/MyDrive/Colab Notebooks/Di Bootcamp files/datascience_salaries.csv'
df = pd.read_csv(file_path)

# Now you can work with the DataFrame 'df'
# For example, to display the first 5 rows:
print(df.head())
```

```python
# prompt: next i want to normailze the salary column

from sklearn.preprocessing import MinMaxScaler

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Fit and transform the 'salary' column
df['salary_normalized'] = scaler.fit_transform(df[['salary']])

# Display the DataFrame with the normalized salary column
print(df.head())
```

```python
from sklearn.decomposition import PCA

# Drop non-numeric columns or encode them before PCA
df_numeric = df.select_dtypes(include='number')

# Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_numeric)

# Convert to DataFrame
pca_df = pd.DataFrame(df_pca, columns=['PC1', 'PC2'])
pca_df.head()
```

```python
# Calculate average and median salary per experience level
grouped = df.groupby('experience_level')['salary'].agg(['mean', 'median']).reset_index()
print(grouped)
```
