```python
df['Age_Group'] = pd.cut(df['Age'], bins=[18, 25, 35, 45, 55, 65],
                         labels=['18-25', '26-35', '36-45', '46-55', '56-65'])

age_group_stats = df.groupby('Age_Group')['treatment'].value_counts(normalize=True).unstack()
print(age_group_stats)
```

```python
gender_stats = df.groupby('Gender')['treatment'].value_counts(normalize=True).unstack()
print(gender_stats)
```

```python
country_stats = df.groupby('Country')['treatment'].value_counts(normalize=True).unstack().fillna(0)
country_stats['Treatment_Rate'] = country_stats.get('Yes', 0)

print("Top 5 countries:\n", country_stats['Treatment_Rate'].sort_values(ascending=False).head())
print("\nBottom 5 countries:\n", country_stats['Treatment_Rate'].sort_values().head())
```
