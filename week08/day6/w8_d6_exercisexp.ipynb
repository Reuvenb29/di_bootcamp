Data visualization is a crucial part of data analysis because it helps us understand complex datasets by presenting them in a visual format. This makes it easier to identify patterns, trends, and outliers that might not be obvious in raw data.

A line graph is a type of data visualization used to display how something changes over time or across a continuous range. It connects data points with lines, making it easy to see trends, increases, decreases, and volatility. It is particularly useful for showing trends in data over time, such as stock prices, temperature changes, or population growth.

# Task
Create a bar chart to visualize the total revenue generated in each quarter of the year (Q1, Q2, Q3, Q4) using the data from "/content/drive/MyDrive/di_bootcamp_resources/week_8/sales_data.csv".

## Convert date column to datetime

### Subtask:
Ensure the 'date' column is in datetime format for easier extraction of the quarter.


**Reasoning**:
Convert the 'date' column to datetime format to extract the quarter information.



```python
df_sales['date'] = pd.to_datetime(df_sales['date'])
display(df_sales.head())
```

## Extract quarter

### Subtask:
Create a new column for the quarter based on the 'date' column.


**Reasoning**:
Create a new column 'quarter' by extracting the quarter from the 'date' column and display the updated DataFrame.



```python
df_sales['quarter'] = df_sales['date'].dt.quarter
display(df_sales.head())
```

## Calculate quarterly revenue

### Subtask:
Group the data by the new 'quarter' column and calculate the sum of 'revenue' for each quarter.


**Reasoning**:
Group the dataframe by quarter and calculate the sum of revenue for each quarter.



```python
quarterly_revenue = df_sales.groupby('quarter')['revenue'].sum()
display(quarterly_revenue)
```

## Create bar chart

### Subtask:
Generate a bar chart using the calculated quarterly revenue, labeling the axes and adding a title.


**Reasoning**:
Generate a bar chart using the calculated quarterly revenue, labeling the axes and adding a title.



```python
import matplotlib.pyplot as plt

plt.bar(quarterly_revenue.index, quarterly_revenue.values)
plt.xlabel("Quarter")
plt.ylabel("Total Revenue ($)")
plt.title("Quarterly Revenue")
plt.show()
```

## Summary:

### Data Analysis Key Findings

* The total revenue for each quarter was calculated: Q1 had \$182,100.0, Q2 had \$183,970.0, Q3 had \$197,680.0, and Q4 had \$194,580.0.
* Q3 generated the highest revenue at \$197,680.0.
* Q1 generated the lowest revenue at \$182,100.0.

```python
import matplotlib.pyplot as plt

# Data for favorite fruits
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
percentages = [40, 30, 20, 10]
colors = ['red', 'yellow', 'purple', 'brown']

# Create the pie chart
plt.figure(figsize=(8, 8)) # Set the size of the pie chart
plt.pie(percentages, labels=fruits, colors=colors, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title("Favorite Fruits Distribution")

# Add a legend
plt.legend(fruits, loc="best")

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Display the chart
plt.show()
```
