```python
# prompt:     Calculate the total quantity of products sold by the company.
#     Identify the category that had the highest revenue and calculate how much revenue it generated.
#     Determine the average revenue per sale in the dataset.
#     Create a bar chart to visualize the total revenue generated in each quarter of the year (Q1, Q2, Q3, Q4).

import pandas as pd
import matplotlib.pyplot as plt
# Calculate the total quantity of products sold
total_quantity = df['Quantity'].sum()
print(f"\nTotal quantity of products sold: {total_quantity}")

# Identify the category with the highest revenue
df['Revenue'] = df['Quantity'] * df['Price']
category_revenue = df.groupby('Category')['Revenue'].sum()
highest_revenue_category = category_revenue.idxmax()
highest_revenue_amount = category_revenue.max()
print(f"Category with the highest revenue: {highest_revenue_category}")
print(f"Revenue generated by {highest_revenue_category}: ${highest_revenue_amount:.2f}")

# Determine the average revenue per sale
average_revenue_per_sale = df['Revenue'].mean()
print(f"Average revenue per sale: ${average_revenue_per_sale:.2f}")

# Create a bar chart to visualize total revenue by quarter
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Quarter'] = df['Order Date'].dt.quarter
quarterly_revenue = df.groupby('Quarter')['Revenue'].sum()

plt.figure(figsize=(8, 5))
plt.bar(quarterly_revenue.index.astype(str), quarterly_revenue.values, color='lightgreen')
plt.title('Total Revenue by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Total Revenue ($)')
plt.show()
```
