```python
# /content/drive/MyDrive/Colab Notebooks/Di Bootcamp files/archive/2019.csv

import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Di Bootcamp files/archive/2019.csv')
df.head()
```

```python
# Check for missing values
print(df.isnull().sum())

# Drop rows with missing values
df = df.dropna()

# Verify that missing values are removed
print(df.isnull().sum())
print(df.info())
```

```python
import matplotlib.pyplot as plt

# Create a figure and axes
fig, ax = plt.subplots(1, 1, figsize=(8, 6))

# Scatter plot of 'social support' vs 'score'
ax.scatter(df['Social support'], df['Score'])
ax.set_xlabel('Social Support')
ax.set_ylabel('Score')
ax.set_title('Relationship between Social Support and Score')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure and axes
fig, ax = plt.subplots(1, 1, figsize=(12, 8))

# Scatter plot of 'GDP per capita' vs 'Healthy life expectancy' by region
sns.scatterplot(data=df, x='GDP per capita', y='Healthy life expectancy', ax=ax)

ax.set_xlabel('GDP per Capita')
ax.set_ylabel('Healthy life expectancy')
ax.set_title('Relationship between GDP per Capita and Healthy life expectancy by Region')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the DataFrame (assuming you've already got it in 'df')
# Sort for cleaner presentation (optional)
df_sorted = df.sort_values(by='GDP per capita', ascending=False)

# Set up the x-axis labels
countries = df_sorted['Country or region']
gdp = df_sorted['GDP per capita']
life_exp = df_sorted['Healthy life expectancy']

# Create figure and axis
fig, ax1 = plt.subplots(figsize=(12, 6))

# Bar chart for GDP per Capita
ax1.bar(countries, gdp, color='lightblue', label='GDP per Capita')
ax1.set_xlabel('Country')
ax1.set_ylabel('GDP per Capita', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_xticklabels(countries, rotation=45, ha='right')

# Line chart for Healthy Life Expectancy on the same x-axis but different y-axis
ax2 = ax1.twinx()
ax2.plot(countries, life_exp, color='green', marker='o', label='Healthy Life Expectancy')
ax2.set_ylabel('Healthy Life Expectancy (Years)', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Title and layout
plt.title('GDP per Capita vs Healthy Life Expectancy by Country')
fig.tight_layout()
plt.show()
```
